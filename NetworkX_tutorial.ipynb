```python
# Import the NetworkX package
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

## Graph
NetworkX provides several classes to store different types of graphs, such as directed and undirected graph. It also provides classes to create multigraphs (both directed and undirected).

For more information, please refer to [NetworkX graph types](https://networkx.org/documentation/stable/reference/classes/index.html).

```python
# Create an undirected graph G
G = nx.Graph()
print(G.is_directed())

# Create a directed graph H
H = nx.DiGraph()
print(H.is_directed())

# Add graph level attribute
G.graph["Name"] = "Bar"
print(G.graph)
```

```python
# Create a random graph with 5 nodes and a probability of 0.5 for edge creation
random_graph = nx.erdos_renyi_graph(5, 0.5)

# Compute the degree matrix
degree_matrix = np.diag([val for (node, val) in random_graph.degree()])
print("Degree Matrix:")
print(degree_matrix)

# Compute the Laplacian matrix
laplacian_matrix = nx.laplacian_matrix(random_graph).toarray()
print("\nLaplacian Matrix:")
print(laplacian_matrix)
```

## Node

Nodes (with attributes) can be easily added to NetworkX graphs.

```python
# Add one node with node level attributes
G.add_node(0, feature=5, label=0)

# Get attributes of the node 0
node_0_attr = G.nodes[0]
print("Node 0 has the attributes {}".format(node_0_attr))
```

```python
G.nodes(data=True)
```

```python
# Add multiple nodes with attributes
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
]) #(node, attrdict)

# Loop through all the nodes
# Set data=True will return node attributes
for node in G.nodes(data=True):
  print(node)

# Get number of nodes
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes))
```

## Edge

Similar to nodes, edges (with attributes) can also be easily added to NetworkX graphs.

```python
# Add one edge with edge weight 0.5
G.add_edge(0, 1, weight=0.5)

# Get attributes of the edge (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Edge (0, 1) has the attributes {}".format(edge_0_1_attr))
```

```python
# Add multiple edges with edge weights
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Loop through all the edges
# Here there is no data=True, so only the edge will be returned
for edge in G.edges():
  print(edge)

# Get number of edges
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
```

## Visualization

```python
# Draw the graph
nx.draw(G, with_labels = True)
```

## Node Degree and Neighbor

```python
node_id = 1

# Degree of node 1
print("Node {} has degree {}".format(node_id, G.degree[node_id]))

# Get neighbor of node 1
for neighbor in G.neighbors(node_id):
  print("Node {} has neighbor {}".format(node_id, neighbor))
```

## Random Graph creation and analysis

```python
# creation of a random network with N nodes and E edges - this a erdos-renyi graph
def create(n,m):
    G = nx.gnm_random_graph(n, m, seed = random_state)
    return G
```

```python
def degree_analysis(G):
    degree_sequence = sorted((d for n, d in G.degree()), reverse=True)

    fig = plt.figure("Degree of a random graph", figsize=(8, 8))
    # Create a gridspec for adding subplots of different sizes
    axgrid = fig.add_gridspec(2, 4)


    ax1 = fig.add_subplot(axgrid[:, :2])
    ax1.plot(degree_sequence, "b-", marker="o")
    ax1.set_title("Degree Rank Plot")
    ax1.set_ylabel("Degree")
    ax1.set_xlabel("Rank")

    ax2 = fig.add_subplot(axgrid[:, 2:])
    ax2.bar(*np.unique(degree_sequence, return_counts=True))
    ax2.set_title("Degree histogram")
    ax2.set_xlabel("Degree")
    ax2.set_ylabel("# of Nodes")

    plt.show()
```

```python
random_state = np.random.RandomState(42)
N=20 #number of nodes
E=2*N #number of edges
G=create(N,E)
```

```python
print(G.edges(nbunch=4))
print(G.adj)
print(G.edges)
print(type(G))
print(len(G.edges))
```

```python
pos_cons = nx.spring_layout(G,seed = random_state)
nx.draw(G, pos=pos_cons,labels ={node:node for node in G.nodes()}) #spring_layout , spectral_layout
plt.draw()
```

```python
# Get the adjacency matrix
A = nx.adjacency_matrix(G).toarray()
print("Adjacency Matrix:")
print(A)



# Get the degree matrix
degree_matrix = np.diag([val for (node, val) in G.degree()])
print('Degree Matrix:')
print(degree_matrix)

# Get the Laplacian matrix
laplacian_matrix = nx.laplacian_matrix(G).toarray()
print('Laplacian Matrix:')
print(laplacian_matrix)
```

## Graph Density and Degree Distribution Analysis
The following analysis computes the graph density and degree distribution for random graphs with varying node sizes.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Function to calculate graph density
def calculate_density(graph):
    num_nodes = graph.number_of_nodes()
    num_edges = graph.number_of_edges()
    if num_nodes > 1:
        density = (2 * num_edges) / (num_nodes * (num_nodes - 1))
        return density
    return 0

# Lists to store density and degree distributions
nodes_list = list(range(20, 201, 20))
densities = []
degree_distributions = []

# Generate random graphs, calculate density and degree distributions
for N in nodes_list:
    random_graph = nx.erdos_renyi_graph(N, 0.5)
    density = calculate_density(random_graph)
    densities.append(density)

    # Degree distribution
    degrees = [degree for node, degree in random_graph.degree()]
    degree_distributions.append(degrees)

    # Plot degree distribution histogram
    plt.figure(figsize=(8, 5))
    plt.hist(degrees, bins=10, edgecolor="black")
    plt.title(f"Degree Distribution for N = {N}")
    plt.xlabel("Degree")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Plot graph density vs number of nodes
plt.figure(figsize=(10, 6))
plt.plot(nodes_list, densities, marker="o")
plt.title("Graph Density vs Number of Nodes")
plt.xlabel("Number of Nodes (N)")
plt.ylabel("Graph Density (D)")
plt.grid(True)
plt.show()
```
